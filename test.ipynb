````python
import re
import json

def extract_json_from_response(response_text):
    # Try to find a JSON code block
    match = re.search(r"```(?:json)?\\s*({[\\s\\S]*?})\\s*```", response_text)
    if match:
        json_str = match.group(1)
    else:
        # Fallback: find the first {...} block
        match = re.search(r"({[\\s\\S]*})", response_text)
        if match:
            json_str = match.group(1)
        else:
            raise ValueError("No JSON object found in response")
    # Now parse the JSON
    print(json_str)
    #return json.loads(json_str)

response_text = 'Based on the provided user input, here is the rewritten format:\n\n```json\n{\n    "rewardType": "Fuel",\n    "spending": 2000,\n    "age": 21,\n    "income": 15000,\n    "employmentType": "salaried"\n}\n```'
start = response_text.find("{")
end   = response_text.rfind("}") + 1

if start == -1 or end == -1:
    raise ValueError("No JSON object found in response_text")

json_block = response_text[start:end]

# 2. Parse and redump
obj = json.loads(json_block)
clean_json_str = json.dumps(obj)            # compact: '{"rewardType":"Fuel",...}'
pretty_json_str = json.dumps(obj, indent=4)  # pretty

print("Compact:", clean_json_str)
print("Pretty:\n", pretty_json_str)
#print(extract_json_from_response(response_text))
````
